# using K neighbors regression

```python
import seaborn as sns  
import pandas as pd  
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Salary_Data.csv")
```

```python
x=dataset.iloc[:,:-1]
y=dataset["Salary"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsRegressor
kn=KNeighborsRegressor(n_neighbors=5)
```

```python
kn.fit(x_train,y_train)
kn.score(x_test,y_test)
kn.score(x_train,y_train)
```

```python
'''to get best accuracy what we can do is get the best value of n_neighbour
that we can get by loop(as we did in classification part of KNN too) , or you can check
to scale the data ,remove the outlier if any present, or
alter the value of random state '''
```

# using decision tree

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)


```

```python


from sklearn.tree import DecisionTreeRegressor,plot_tree
dt=DecisionTreeRegressor() # it doesn't have criterion as those in classifications, like entropy or gini,;here are squared_error etc..
dt.fit(x_train,y_train)
dt.score(x_test,y_test)


```

```python
plot_tree(dt)
plt.show()
```

```python
# can check whether the model is overfitting or not and can try pruning
```
